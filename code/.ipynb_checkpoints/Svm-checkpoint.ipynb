```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../dataset/train.csv")
data
```

```python
data.describe()
```

```python
data=data.dropna(how='any')
data.describe()
```

```python
plt.subplots(figsize=(8, 8))
sns.heatmap(data.corr().round(2),annot=True)
```

# 划分数据集

```python
labels=np.array(data['Transported'])
features=data.drop('Transported',axis=1)
```

```python
features=features.drop('PassengerId',axis=1)
features=features.drop('Name',axis=1)
features=features.drop('Cabin',axis=1)
features = pd.get_dummies(features)
features_list=list(features.columns)
features.head(5)
```

```python
features=np.array(features)
```

```python
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels=train_test_split(features,labels,test_size=0.25,random_state=42)
train_features.shape
```

```python
from sklearn import svm
clf = svm.SVC() #  默认kernel='rbf'
clf = clf.fit(train_features, train_labels)
```

```python
prediction = clf.predict(test_features)
prediction
```

```python
import Utils
# 获取混淆矩阵
confusion = Utils.getConfusion(test_labels, prediction)
confusion
```
